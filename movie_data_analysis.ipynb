```python
def collect_json(column_name, json_key):
    keywords = []
    for n in range(0, df.shape[0]):
        array = []
        a = json.loads(df[column_name].iloc[n])
        for i in a:
            b = i[json_key]
            array.append(b)
        keywords.append(array)
    return keywords

def data_df(keywords):
    new_data = []
    for i in keywords:
        array_vals = []
        data = ",".join(i)
        array_vals.append(data)
        new_data.append(array_vals)
    return new_data

def dataframe(new_data, new_col_name):
    new_df = pd.DataFrame(new_data)
    new_df.columns = [new_col_name]
    return new_df
```

```python
y_df = collect_json("keywords", "name")
df_data = data_df(y_df)
final_df = dataframe(df_data, "all_key_words")
final_df
```

```python
y_df = collect_json("spoken_languages","iso_639_1")
df_data = data_df(y_df)
df_lang = dataframe(df_data, "language")
df_lang
```
